## Baseline Model - Logical Regression

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score


# Load the data
train_data = pd.read_csv('TMDB_train.csv')
test_data = pd.read_csv('TMDB_evaluate.csv')
final_train_data = pd.read_csv('TMDB_test.csv')

# Identifying the numerical columns 
numerical_cols = train_data.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_cols.remove('rate_category') 
numerical_cols.remove('average_rate')  

X_train = train_data[numerical_cols]
y_train = train_data['rate_category']
X_test = test_data[numerical_cols]
y_test = test_data['rate_category']

X_final_train = final_train_data[numerical_cols]

# Standardizing the data
scaler = StandardScaler()  
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_final_train_scaled = scaler.transform(X_final_train)

# Initialize and train Logistic Regression
model = LogisticRegression()
model.fit(X_train_scaled, y_train)

# Evaluate the model on the training set
train_predictions = model.predict(X_train_scaled)
train_accuracy = accuracy_score(y_train, train_predictions)
print(f'Train Accuracy: {train_accuracy}')

# Evaluate the model on the TMDB_evaluate set
test_predictions = model.predict(X_test_scaled)
test_accuracy = accuracy_score(y_test, test_predictions)
print(f'Test Accuracy: {test_accuracy}')

# Final prediction on TMDB_test 
final_test_prediction = model.predict(X_final_train_scaled)

# Creating a CSV for Kaggle
results_df = pd.DataFrame({
    'id': final_train_data['id'], 
    'rate_category': final_test_prediction})

# Saving to a CSV file
results_filename = 'final_test_predicted_rate_category_LR.csv'
results_df.to_csv(results_filename, index=False)
```

```python
################ Optimisation 1/1 - Baseline - Logical Regression - Best iteration number ######################

# Setting up GridSearchCV
params = {'max_iter': [100, 200, 300, 400, 500, 1000] 
         }
model = LogisticRegression()
clf = GridSearchCV(model, params, cv=3)
clf.fit(X_train_scaled, y_train)

# Best parameter
print("Best parameters found: ", clf.best_params_['max_iter'])
#print("Best accuracy found: ", clf.best_score_)

# Evaluate on the test set with the best parameters
best_model = clf.best_estimator_
predictions = best_model.predict(X_test_scaled)
test_accuracy = accuracy_score(y_test, predictions)
print(f'Test Accuracy with Optimal max_iter: {test_accuracy}')
```

```python
################ Evaluation 1/2 - Baseline - Logical Regression - Confusion Matrix ######################

from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
disp.ax_.set_title("Confusion Matrix for Logistic Regression")
print("Confusion Matrix:\n", cm)
```

```python
################ Evaluation 2/2 - Baseline - Logical Regression - Bias and Variance ######################
from mlxtend.evaluate import bias_variance_decomp

X_train_scaled = X_train_scaled.to_numpy() if hasattr(X_train_scaled, 'to_numpy') else X_train_scaled
y_train = y_train.to_numpy() if hasattr(y_train, 'to_numpy') else y_train
X_test_scaled = X_test_scaled.to_numpy() if hasattr(X_test_scaled, 'to_numpy') else X_test_scaled
y_test = y_test.to_numpy() if hasattr(y_test, 'to_numpy') else y_test

model = LogisticRegression(max_iter = 50)

# Estimate bias and variance
mse, bias, var = bias_variance_decomp(
    model, X_train_scaled, y_train, X_test_scaled, y_test, loss='0-1_loss', random_seed=123)

print('Average expected loss: %.3f' % mse)
print('Average bias: %.3f' % bias)
print('Average variance: %.3f' % var)
```

## Hypothesis
**Main Hypothesis:** 

"The use of unlabeled data through semi-supervised learning techniques significantly improves the predictive performance of models on the TMDB dataset compared to using only labeled data."

**How to test hypothesis:**

We will employ a comparative analysis approach. This involves training a baseline supervised model exclusively on the labeled dataset and then training a semi-supervised model utilizing both labeled and unlabeled datasets. The performance of both models will be evaluated and compared using accuracy, precision, recall, and F1-score metrics on a separate validation set to ascertain the efficacy of incorporating unlabeled data in enhancing model predictions.

## ML Model 1: Multinomial NB

```python
import numpy as np
import pandas as pd
from scipy.sparse import load_npz
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV

# Load training data
train_features = load_npz('TMDB_text_features_tfidf/train_concat_tfidf.npz')
train_labels = pd.read_csv('TMDB_train.csv')['rate_category']

# Load evaluation data
eval_features = load_npz('TMDB_text_features_tfidf/eval_concat_tfidf.npz')
eval_labels = pd.read_csv('TMDB_evaluate.csv')['rate_category']

# Load final training data
final_train_features = load_npz('TMDB_text_features_tfidf/test_concat_tfidf.npz')
final_train_data = pd.read_csv('TMDB_test.csv')

# Train a MNB model
nb_model = MultinomialNB()
nb_model.fit(train_features, train_labels)

# Predict on the evaluation set
eval_predictions = nb_model.predict(eval_features)

# Evaluate the model
accuracy = accuracy_score(eval_labels, eval_predictions)
print(f"Accuracy on evaluation set: {accuracy}")
#print(classification_report(eval_labels, eval_predictions))

# Final prediction on TMDB_test 
final_test_prediction = nb_model.predict(final_train_features)

# Creating a CSV for Kaggle
results_df = pd.DataFrame({
    'id': final_train_data['id'],  # Ensure 'id' column is correctly named in your CSV
    'rate_category': final_test_prediction
})

# Saving to a CSV file
results_filename = 'final_test_predicted_rate_category_MNB.csv'
results_df.to_csv(results_filename, index=False)
```

```python
######################## Optimisation 1/1 - ML Model 1- MNB - Grid Search for Smooting Parameter Alpha #############################

param_grid = {
    'alpha': [0.01, 0.1, 0.5, 1.0, 10.0],  
}

nb = MultinomialNB()
grid_search = GridSearchCV(estimator=nb, param_grid=param_grid, cv=5, scoring='accuracy')

grid_search.fit(train_features, train_labels)

print("Best parameters:", grid_search.best_params_)
print("Best cross-validation accuracy: {:.3f}".format(grid_search.best_score_))

best_nb = grid_search.best_estimator_
nb_predictions = best_nb.predict(eval_features)
accuracy = accuracy_score(eval_labels, nb_predictions)
print(f"Accuracy on the evaluation set: {accuracy}")

```

```python
######################## EVALUATION 1/2 - MNB - Confusion Matrix #############################
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(eval_labels, nb_predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix for Multinomial Naive Bayes')
print("Confusion Matrix:\n", cm)
plt.show()
```

```python
######################## EVALUATION 2/2 - MNB - Bias and Variance #############################
from mlxtend.evaluate import bias_variance_decomp
from scipy.sparse import load_npz
from sklearn.naive_bayes import MultinomialNB


train_features = load_npz('TMDB_text_features_tfidf/train_concat_tfidf.npz')
train_labels = pd.read_csv('TMDB_train.csv')['rate_category'].values  

eval_features = load_npz('TMDB_text_features_tfidf/eval_concat_tfidf.npz')
eval_labels = pd.read_csv('TMDB_evaluate.csv')['rate_category'].values  

nb_model = MultinomialNB()

# Converting sparse matrices to dense, as mlxtend's function does not support sparse input
X_train_dense = train_features.toarray()
X_eval_dense = eval_features.toarray()

mse, bias, var = bias_variance_decomp(
    nb_model, X_train_dense, train_labels, X_eval_dense, eval_labels,
    loss='0-1_loss', random_seed=123)

print('Mean Squared Error:', mse)
print('Bias:', bias)
print('Variance:', var)
```

## ML Model 2: K-Nearest Neighbour


```python
import numpy as np
import pandas as pd
from scipy.sparse import load_npz
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
from sklearn import datasets


# Load training data
train_features = load_npz('TMDB_text_features_tfidf/train_concat_tfidf.npz')
train_labels = pd.read_csv('TMDB_train.csv')['rate_category']

# Load evaluation data
eval_features = load_npz('TMDB_text_features_tfidf/eval_concat_tfidf.npz')
eval_labels = pd.read_csv('TMDB_evaluate.csv')['rate_category']

# Load final training data
final_train_features = load_npz('TMDB_text_features_tfidf/test_concat_tfidf.npz')
final_train_data = pd.read_csv('TMDB_test.csv')

knn_model = KNeighborsClassifier(n_neighbors=5)  

knn_model.fit(train_features, train_labels)

# Predict on the evaluation set
eval_predictions = knn_model.predict(eval_features)

# Evaluate the model on evaluation set
accuracy = accuracy_score(eval_labels, eval_predictions)
print(f'Accuracy on evaluations: {accuracy}')

# Final prediction on TMDB_test 
final_test_prediction = knn_model.predict(final_train_features)

# Creating a CSV for Kaggle
results_df = pd.DataFrame({
    'id': final_train_data['id'],  # Ensure 'id' column is correctly named in your CSV
    'rate_category': final_test_prediction
})

# Saving to a CSV file
results_filename = 'final_test_predicted_rate_category_KNN.csv'
results_df.to_csv(results_filename, index=False)
```

```python
######################## OPTIMISATION 1/1 - KNN - Changing K #############################

k_list = [3,5,10,20,30,100]

for k in k_list:
    knn_model = KNeighborsClassifier(n_neighbors=k)  # You can tune the number of neighbors
    
    # Train the model
    knn_model.fit(train_features, train_labels)
    
    # Predict on the evaluation set
    eval_predictions = knn_model.predict(eval_features)
    
    # Evaluate the model
    accuracy = accuracy_score(eval_labels, eval_predictions)
    print("\nKNN with k =", k ," score is %f " %accuracy)
#print(f'Accuracy: {accuracy}')
#print(classification_report(eval_labels, eval_predictions))
```
